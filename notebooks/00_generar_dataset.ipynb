# Generar dataset

Referencia para ejemplo: maderea (proveedores de madera estructual)
https://www.maderea.es/proveedores-de-madera-estructural/

```python
import pandas as pd
import numpy as np
import random
```

```python

np.random.seed(42)

n = 300  # número de filas

clientes = ["Construcciones Norte", "Promociones Sur", "Reformas López", "Obras y Casas", "Muebles Pérez Polo"]
productos = ["CLT", "Viga laminada", "Viga"]
maderas = ["Pino", "Abeto", "PINO", "abeto"]
certificaciones = ["CE", "FSC", "PEFC", None]
estados = ["cerrado", "CERRADO", "Cerrada", "ok", "pendiente"]
paises = ["España", "ES", "Madrid"]

data = []

for i in range(n):
    row = {
        " ID Venta ": 1000 + i,
        "cliente": random.choice(clientes) if random.random() > 0.2 else None,
        "Cliente": random.choice(clientes) if random.random() > 0.7 else None,
        "fecha venta": random.choice(["01/02/2024", "2024-02-03", "03-02-2024", "2024/02/05"]),
        "Fecha_Venta": None if random.random() > 0.5 else "2024-02-06",
        "producto ": random.choice(productos),
        "tipo_madera": random.choice(maderas),
        "certificacion": random.choice(certificaciones),
        "cantidad_m3 ": random.choice([0, 1.5, 2.0, 2.5, None]),
        "precio_m3": random.choice([320, 450]),
        "importe ": None,
        "estado": random.choice(estados),
        "comercial": random.choice(["Juan", "Ana"]),
        "pais": random.choice(paises)
    }
    
    # calcular importe si hay datos
    if row["cantidad_m3 "] and row["precio_m3"]:
        row["importe "] = row["cantidad_m3 "] * row["precio_m3"]
    
    data.append(row)

df_dirty = pd.DataFrame(data)

# añadir duplicados
df_dirty = pd.concat([df_dirty, df_dirty.sample(20)])

df_dirty.head()
```

```python
df_dirty.to_excel("../data/raw/dirty_sales_data.xlsx", index=False)
```
